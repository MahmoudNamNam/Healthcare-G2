# **Installation Guide**  

Before running the chatbot, ensure that all required dependencies are installed.  

---

## **1. Install Python (if not installed)**  

Make sure you have **Python 3.8 or later** installed. You can check your Python version by running:  

```sh
    python --version
```  

If Python is not installed, download and install it from [Python’s official website](https://www.python.org/downloads/).  

---

## **2. Install Required Libraries**  

Install the necessary Python packages using `pip`:  

```sh
    pip install -r requirements.txt
```  

---

## **3. Install Ollama**  

Ollama is required to run the language model locally. You can download and install it from:  

🔗 [Ollama Official Website](https://ollama.com/)  

Once installed, verify that it is running locally by executing:  

```sh
    ollama list
```  

Then, pull the required model by running:  

```sh
    ollama pull Koesn/llama3-8b-instruct
```  

---

# **How to Run the Chatbot**  

### **1️⃣ Running with Streamlit (Web Interface)**  

Once all dependencies are installed, you can start the chatbot in your browser by running:  

```sh
    streamlit run app.py
```  

This will launch the chatbot in your default web browser.  

---

### **2️⃣ Running in the Terminal (Command-Line Interface)**  

If you prefer running the chatbot in the terminal, use the following steps:  

1. Save the chatbot script as `chatbot.py` (if not already done).
2. Run the chatbot with the command:

```sh
    python chatbot.py
```

3. Start chatting! Type your psychology-related questions and enter `"exit"` to quit.
